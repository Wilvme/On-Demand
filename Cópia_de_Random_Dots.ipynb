```python
import numpy as np, matplotlib.pyplot as plt
x= np.random.rand(200); y = np.random.rand(200)
colors = np.random.rand(200)
plt.scatter(x, y, c=colors, cmap='rainbow', s=30)
plt.axis('off'); plt.show()


```
